```python
import os
import cv2
import matplotlib.pyplot as plt

images_path = '/content/drive/MyDrive/cartoons'
output_dir = '/content/drive/MyDrive/cartoons'

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate over each image file in the images_path directory
for filename in os.listdir(images_path):
    image_path = os.path.join(images_path, filename)

    # Read the grayscale image
    gray_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply binary thresholding
    _, binary_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)

    # Save the binary image to the output directory
    output_path = os.path.join(output_dir, f'binary_{filename}')
    cv2.imwrite(output_path, binary_img)

    # Display the original and binary images
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(binary_img, cmap='gray')
    plt.title('Binary Image')
    plt.axis('off')

    plt.show()
```
