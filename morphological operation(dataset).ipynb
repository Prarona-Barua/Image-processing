```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

input_folder = '/content/drive/MyDrive/binary image'
output_folder = '/content/drive/MyDrive/binary image'

os.makedirs(output_folder, exist_ok=True)

# Define the kernel for dilation and erosion
kernel = np.ones((5, 5), np.uint8)

# Iterate over the images in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Adjust the file extensions as needed
        # Read the image
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)

        # Convert the image to grayscale
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply morphological operations
        dilated = cv2.dilate(gray_image, kernel, iterations=2)
        eroded = cv2.erode(gray_image, kernel, iterations=2)

        # Save the morphological operation results in the output folder
        cv2.imwrite(os.path.join(output_folder, f'dilated_{filename}'), dilated)
        cv2.imwrite(os.path.join(output_folder, f'eroded_{filename}'), eroded)

# Iterate over the files in the output folder
for filename in os.listdir(output_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Adjust the file extensions as needed
        # Read the image
        image_path = os.path.join(output_folder, filename)
        image = cv2.imread(image_path)

        # Convert the image from BGR to RGB for proper display
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Display the image using matplotlib
        plt.imshow(image_rgb)
        plt.title(filename)  # Set the title as the filename
        plt.axis('off')  # Turn off axis labels
        plt.show()
```
